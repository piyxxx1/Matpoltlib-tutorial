```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie([1])
plt.show()
```

```python
#ploting data
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students , labels = classes)
plt.show()
```

```python
#slice parameter 
explode = [0,0,0.2,0,0]
plt.pie(class1_students , labels = classes , explode = explode)
plt.show()
```

```python
#color parameter
colors = ['c' , 'm' , 'r' , 'y' ,'g']
plt.pie(class1_students , labels = classes , colors = colors)
plt.show()
```

```python
#showes the percentage
colors = ['c' , 'm' , 'r' , 'y' ,'g']
plt.pie(class1_students , labels = classes , colors = colors, autopct = '%0.1f%%' )
plt.show()
```

```python
#sahdow
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True)
plt.show()
```

```python
#radius
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True , radius = 1.4)
plt.show()
```

```python
# rotate
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True , radius = 1.4 , startangle= 180)
plt.show()
```

```python
#fontsize
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
textprops = {'fontsize' : 12}
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True , textprops = textprops)
plt.show()
```

```python
#counterclock
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
textprops = {'fontsize' : 12}
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True , textprops = textprops , counterclock = False)
plt.show()
```

```python
#linewidth , width , edgecolor
wedgeprops = {"linewidth": 1, 'width':1, "edgecolor":"k"}
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%' , shadow =True , textprops = textprops ,wedgeprops=wedgeprops) 
plt.show()
```

```python
plt.figure(figsize = (18,10))
colors = ['c' , 'm' , 'r' , 'y' ,'g']
explode = [0,0,0.2,0,0]
textprops = {'fontsize' : 12}
plt.pie(class1_students , labels = classes , explode = explode, autopct = '%0.1f%%'  , textprops = textprops , counterclock = False ,wedgeprops=wedgeprops)
plt.legend()
plt.show()
```
