```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45 , (100))
py_student_age = np.random.randint(15,40 , (100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
#plot a histogram
plt.title('ML Student age histogram',fontsize = 15)
plt.ylabel('Quantity' , fontsize = 13)
plt.xlabel('Age catagory' , fontsize = 13)
plt.hist(ml_student_age)
plt.show()
```

```python
#set your own distance using bins parameter
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize = 15)
plt.ylabel('Quantity' , fontsize = 13)
plt.xlabel('Age catagory' , fontsize = 13)
plt.hist(ml_student_age , bins )
plt.show()
```

```python
#setting your own width using rwidth parameter
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize = 15)
plt.ylabel('Quantity' , fontsize = 13)
plt.xlabel('Age catagory' , fontsize = 13)
plt.hist(ml_student_age , bins , rwidth=0.8 )
plt.show()
```

```python
# set your own hist type either bar or step etc
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist(ml_student_age , bins , rwidth=0.8 , histtype="step")
plt.show()
```

```python
# set horizontal or vertical
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist(ml_student_age , bins , rwidth=0.8 ,orientation='horizontal')
plt.show()
```

```python
#bar color pluse boder color
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist(ml_student_age , bins , rwidth=0.8 ,orientation='horizontal' , color = 'grey' ,edgecolor='black')
plt.show()
```

```python
# legend
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist(ml_student_age , bins , rwidth=0.8 ,orientation='horizontal' , label = 'ML Student')
plt.legend()
plt.show()
```

```python
# plot both the data x , y
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist([ml_student_age ,py_student_age] , bins , rwidth=0.8 ,orientation='vertical' , color =['grey', 'red'] ,edgecolor='black',label = ['ML Student' , 'PY Student'])
plt.legend()
plt.show()
```

```python
#increase size of graph using figure function
bins = [15 , 20 , 25 , 30 , 35 , 40 , 45]
plt.figure(figsize=(16,9))
plt.title('ML Student age histogram',fontsize=15)
plt.ylabel('Quantity' , fontsize=13)
plt.xlabel('Age catagory' , fontsize=13)
plt.hist([ml_student_age ,py_student_age] , bins , rwidth=0.8 ,orientation='vertical' , color =['grey', 'red'] ,edgecolor='black',label = ['ML Student' , 'PY Student'])
plt.legend()
plt.show()
```
