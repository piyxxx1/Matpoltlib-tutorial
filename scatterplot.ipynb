```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_google = pd.read_csv('googleplaystore.csv' , nrows=10000)
df_google.shape # shape is a numpy attribute to see row and cloumne
df_google
```

```python
x=df_google['Rating']
y=df_google['Reviews']
```

```python
plt.scatter(x , y )
```

```python
plt.scatter(x , y)
plt.title('Google play store app scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x , y , color = 'red')
plt.title('Google play store app scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x , y , color = 'red' , marker ='*' , s =100)
plt.title('Google play store app scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x , y , color = 'red' , marker ='*' , s =100 , alpha = 0.1)
plt.title('Google play store app scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
#double plot
plt.figure(figsize=(16,9))
plt.scatter(x , y , color = 'red' , marker ='*' , s =100 , alpha = 0.1)
plt.scatter(x , df_google['Installs'] , color = 'm' , marker ='o' , s =100 , alpha = 0.1)
plt.title('Google play store app scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews & install')
```
