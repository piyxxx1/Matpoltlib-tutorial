```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = np.array(["Python", "R", "AI", "ML", "DS"])
class1_students = np.array([30, 10, 20, 25, 10]) # out of 100 student in each class
class2_students = np.array([40, 5, 20, 20, 10])
class3_students = np.array([35, 5, 30, 15, 15])
```

```python
plt.bar(classes,class1_students ,  color = 'b' ,edgecolor = 'black' )
```

```python
#to change width 
plt.bar(classes,class1_students ,  color = 'b' ,edgecolor = 'black' , width = 0.2 )
```

```python
# for changing alignment
plt.bar(classes,class1_students ,  color = 'r' ,edgecolor = 'black'  , align = 'edge')
```

```python
#opacity
plt.bar(classes,class1_students ,  color = 'r' ,edgecolor = 'black'  , alpha = 0.5)
```

```python
#change the line style
plt.bar(classes,class1_students ,  color = 'r' ,edgecolor = 'black'  ,linestyle = '--')
```

```python
#direction
style.use('ggplot')
plt.barh(classes,class1_students ,  color = 'r' ,edgecolor = 'black' ,linestyle = '--' , label = 'class 1 student')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students ,  color = 'r' ,edgecolor = 'black' ,linestyle = '--' , label = 'class 1 student')
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students , width=0.3, color = 'y' ,edgecolor = 'black' ,linestyle = '--' , label = 'class 1 student')
plt.legend()
plt.xlabel('job roles' , fontsize=25)
plt.ylabel('total size',  fontsize=25)
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index,class1_students , width, color = 'r' ,edgecolor = 'black' ,label = 'class 1 student')
plt.bar(classes_index+width,class2_students ,width,  color = 'b' ,edgecolor = 'black' ,label = 'class 2 student')
plt.bar(classes_index+width+width,class3_students , width, color = 'y' ,edgecolor = 'black' ,label = 'class 3 student')
plt.legend()
plt.xticks(classes_index+width,classes)
plt.ylabel('Total size', fontsize=25)
plt.show()
```

```python
#rotation of x side 
classes_index = np.arange(len(classes))
 
width = 0.2
 
plt.bar(classes_index, class1_students, width , color = "b",label =" Class 1 Students") #visible=False
plt.bar(classes_index + width, class2_students, width , color = "g",label =" Class 2 Students") 
plt.bar(classes_index + width + width, class3_students, width , color = "y",label =" Class 2 Students") 
 
plt.xticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart of IAIP Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

```python
#horizontal 
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width = 0.2
plt.barh(classes_index,class1_students , width, color = 'r' ,edgecolor = 'black' ,label = 'class 1 student')
plt.barh(classes_index+width,class2_students ,width,  color = 'b' ,edgecolor = 'black' ,label = 'class 2 student')
plt.barh(classes_index+width+width,class3_students , width, color = 'y' ,edgecolor = 'black' ,label = 'class 3 student')
plt.legend()
plt.yticks(classes_index+width,classes)
plt.xlabel('Total size', fontsize=25)
plt.show()
```
